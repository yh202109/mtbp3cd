## Example - 1

Generate data summary tables. The output includes 3 df:

1. Count of rows of input data by output-col-columns and output-row-columns
2. Normalized counts by perct_within_index
3. Report with format 

To create a dataset:

```python
data = {
    'A': ['foo','foo', 'foo', 'foo', 'foo', 'foo', 'bar', 'bar', 'baz', 'baz', 'baz'],
    'B': ['one','one','one','one', 'one', 'two', 'two', 'one', 'one', 'two', 'two'],
    'C': ['y','x','x','x', 'y', 'x', 'y', 'x', 'y', 'x', 'y'],
    'D': ['apple','apple','apple','apple', 'banana', 'apple', 'banana', 'apple', 'banana', 'apple', 'banana'],
    'E': ['red','blue','red','red', 'red', 'blue', 'blue', 'red', 'blue', 'red', 'blue'],
    'value': [0,1,1,1, 2, 3, 4, 5, 6, 7, 8]
}
df = pd.DataFrame(data)

# Use crosstab_from_lists with 3-level multi-index for rows and columns
rows = ['A', 'B', 'C']
cols = ['D', 'E']
```

Create a frequency (count) table:

```python
(crosstab_from_lists(df, rows, cols))['report']
```

Create a table with row-wise percentage:

```python
(crosstab_from_lists(df, rows, cols, rows))['report']
```

Create a table with grouped row-wise percentage:

```python
(crosstab_from_lists(df, rows, cols, ['A','B']))['report']
```

Create a table with column-wise percentage:

```python
(crosstab_from_lists(df, rows, cols, cols))['report']
```

Create a table with grouped column-wise percentage:

```python
(crosstab_from_lists(df, rows, cols, ['D']))['report']
```

Create a table with grouped row-and-column-wise percentage:

```python
(crosstab_from_lists(df, rows, cols, ['A', 'D']))['report']
```
